# First-order reaction in isothermal spherical catalyst pellet

Effect of Thiele modulus on concentration profile in the isothermal spherical pellet for fixed reactant concentration at the pellet external surface.

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np

#  specify values of Thiele moduli
phi_1 = 1.0 
phi_2 = 2.0 
phi_3 = 5.0

plt.rc('font', family='serif')
plt.rc('figure',figsize=(5, 4))
plt.rc('text', usetex=True)

rho = np.linspace(1.e-5, 1, 501)   
c = np.zeros(len(rho))

#  calculate and plot profiles  
c = (1/rho)*np.sinh(phi_1*rho)/np.sinh(phi_1)
plt.plot(rho, c, '--', linewidth=1.5, color='b')

c = (1/rho)*np.sinh(phi_2*rho)/np.sinh(phi_2) 
plt.plot(rho, c, '-', linewidth=1.5, color='r')

c = (1/rho)*np.sinh(phi_3*rho)/np.sinh(phi_3) 
plt.plot(rho, c, '-.', linewidth=1.5, color='k') 

#  specify plot axes, title and legend
plt.xlabel(r'Dimensionless radius, $\rho = r/R$ [-]')
plt.ylabel(r'Dimensionless concentration, $c = C_A/C_{As}$ [-]')
plt.title('First-order reaction in isothermal spherical pellet as a function of Theile Moduli') 

plt.legend([(r'$\phi_1 =  '+str(phi_1)+'$'),(r'$\phi_2 =  '+str(phi_2)+'$'),
            (r'$\phi_3 =  '+str(phi_3)+'$')],loc='lower right')

plt.axis([0.,1.,0.0,1.0])

plt.show()
```
