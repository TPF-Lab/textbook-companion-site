# First-order chemical reaction in isothermal spherical catalyst pellet

The dimensionless mass balance equation for the first-order chemical reaction in the isothermal spherical catalyst pellet is

$$
\frac{d^2 c }{d \rho ^2}  + \frac{2}{\rho} \cdot \frac{d c }{d \rho} - \phi^2 \cdot c = 0
$$

The boundary conditions  are

- at the particle center, $\rho =0$:

$$
\frac{d c}{d \rho} =0 
$$

- at the outer particle surface,  $\rho =1$:

$$
1 - c\;\; = \;\;\frac{1}{{B{i_m}}} \cdot {\left. {\frac{{dc}}{{d\rho }}} \right|_{\rho  = 1}}
$$  

The analytical solution of the second-order differential equation is 

$$c = \frac{1}{\rho } \cdot \frac{{\sinh (\phi  \cdot \rho )}}{{\sinh (\phi ) + \frac{{\phi  \cdot \cosh (\phi ) - \sinh (\phi )}}{{B{i_m}}}}}$$

```python
%matplotlib inline
import numpy as np
from matplotlib.pylab import *
from IPython.display import display
from ipywidgets import interact, FloatSlider, HTML, HTMLMath
#
# Main function
#
def fmain (phi, Bi1, Bi2, Bi3):
    """Calculate and plot the dimensionless reactant concentration
    
    Calculate and plot dimensionless reactant concentration 
    profiles in the radial direction of a spherical pellet
    for various Biot numbers

    Parameters
    ----------
    phi: float
        Theile Modulus

    Bi1, Bi2, Bi3: float
        Biot numbers
    """
    #  specify the array of equally distributed radial positions
    rho = np.linspace(1.e-5, 1, 501)

    #  allocate array of concentrations     
    c = np.zeros(len(rho))    

    #  calculate and plot profile for Bi1
    c = (1./rho)*np.sinh(phi*rho)/(np.sinh(phi) + (phi*np.cosh(phi)- np.sinh(phi))/Bi1)
    matplotlib.rcParams.update({'font.size': 12})    
    plot(rho, c, '-',  linewidth=1.5, color='b')

    #  calculate and plot profile for Bi2
    c = (1./rho)*np.sinh(phi*rho)/(np.sinh(phi) + (phi*np.cosh(phi)- np.sinh(phi))/Bi2)
    plot(rho, c, '-', linewidth=1.5, color='r')
    
    #  calculate and plot profile for Bi3
    c = (1./rho)*np.sinh(phi*rho)/(np.sinh(phi) + (phi*np.cosh(phi)- np.sinh(phi))/Bi3) 
    plot(rho, c, '-', linewidth=1.5, color='k')
    
    #  specify plot axes, title and legend
    xlabel(r'Dimensionless radius, $\rho$ [-]')
    ylabel('Dimensionless concentration, $c $ [-]')
    title1 = 'First-order reaction in isothermal spherical pellet\n with external mass transfer limitations, $\phi=%s$' 
    title (title1 % (phi))
    legend([(r'${Bi} = '+str(Bi1)+'$'),(r'${Bi} = '+str(round(Bi2,2))+'$'),
            (r'${Bi} = '+str(Bi3)+'$')],loc='lower right')    
    axis([0.,1.,0.0,1.0])
    show()
    return

# specify widgets
h1=HTML(value="<h4><b> First-Order Reaction in Isothermal Spherical Pellet <br> with External Mass Transfer Limitations</b>")
h2 = HTMLMath(value=r"<h4>Specify Thiele modulus $\phi$ and three values of Biot number $Bi$ </br> to compare concentration profiles:")
display(h1)
display(h2)
# Use slider to input the value of Thiele modulus
style = {'description_width': 'initial'}
layout={'width': '400px'}
phi_slider = FloatSlider(min=0.5, max=10, step=0.5, value=2, style=style, layout=layout)
phi_slider.description = r'Theile'
phi_slider.style.handle_color = 'green'
# Use slider to input values of Biot number
Bi1_slider = FloatSlider(min=0.5, max=10, step=0.5, value=1, style=style, layout=layout)
Bi1_slider.description = r'Bi_1'
Bi1_slider.style.handle_color = 'blue'
Bi2_slider = FloatSlider(min=0.5, max=10, step=0.5, value=3, style=style, layout=layout)
Bi2_slider.description = r'Bi_2'
Bi2_slider.style.handle_color = 'red'
Bi3_slider = FloatSlider(min=0.5, max=10, step=0.5, value=10, style=style, layout=layout)
Bi3_slider.description = r'Bi_3'
Bi3_slider.style.handle_color = 'black'
w = interact(fmain, phi = phi_slider, Bi1 = Bi1_slider, Bi2 = Bi2_slider, Bi3 = Bi3_slider)
```
